```python
import pandas as pd

# Load Indian dataset
df1 = pd.read_csv('data/indian_startup_funding_2020_2025_sample.csv')
print("Indian Dataset:")
print(df1.shape)
print(df1.columns.tolist())
```

```python
# Load Global Crunchbase dataset
df2 = pd.read_csv('data/big_startup_secsees_dataset.csv')
print("Global Dataset:")
print(df2.shape)
print(df2.columns.tolist())
```

```python
# Check status column values
print(df2['status'].value_counts())
print("\nCountry distribution:")
print(df2['country_code'].value_counts().head(10))
```

```python
# Filter Indian startups from global dataset
df2_india = df2[df2['country_code'] == 'IND']
print("Indian startups in global dataset:")
print(df2_india.shape)
print(df2_india['status'].value_counts())
```

```python
# Prepare df2_india for merging
# Rename columns to match df1
df2_india_clean = df2_india[[
    'name', 'category_list', 'funding_total_usd',
    'status', 'country_code', 'city',
    'funding_rounds', 'founded_at',
    'first_funding_at', 'last_funding_at'
]].copy()

# Rename columns
df2_india_clean.columns = [
    'Startup', 'Industry', 'TotalFunding',
    'Status', 'Country', 'City',
    'FundingRounds', 'FoundedAt',
    'FirstFundingAt', 'LastFundingAt'
]

print("Cleaned Indian dataset:")
print(df2_india_clean.shape)
print(df2_india_clean.head())
```
